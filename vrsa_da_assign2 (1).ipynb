# K-Means Clustering 

```python
#importing necessary pakages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
import seaborn as sns
from scipy.spatial import distance_matrix
from sklearn.cluster import KMeans
```

```python
num_var2 = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustSpeed', 'WindSpeed', 'Humidity', 'Pressure','Cloud','Temp']
```

```python
df2 = newdf[num_var2]
df2.head()
```

```python
# scaling the the numerical values in every numerical field
min_max_scaler = preprocessing.MinMaxScaler()
np_scaled = min_max_scaler.fit_transform(df2)
df_normalized = pd.DataFrame(np_scaled)
df_normalized.columns = num_var2
df_normalized.describe()
```

```python
# caluculating sse for clusters from 1 to 10
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df_normalized)
    sse.append(km.inertia_)
```

```python
#plotting elbow curve
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2)
kmeans.fit(df_normalized)
df_clusters = df_normalized.copy()
df_clusters["cluster"] = kmeans.fit_predict(df_normalized)
df_clusters.head()
```

```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import os
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
import sklearn.metrics
```

```python
pred_train, pred_test, tar_train, tar_test = train_test_split(predictors, targets, test_size=.4)
```

```python
# find the best tree depth for the model

for iteration in range(15):
    acc_list = list()
    for i in range(1,15):
        classifier=DecisionTreeClassifier(max_depth=i)
        classifier=classifier.fit(pred_train,tar_train)

        predictions=classifier.predict(pred_test)

        accuracy = sklearn.metrics.accuracy_score(tar_test, predictions)
        acc_list.append(accuracy)

print (acc_list)
```

```python
print ("Best depth for the classifier is:", acc_list.index(max(acc_list)))
```

```python
# Build model on training data
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

classifier=DecisionTreeClassifier(max_depth=7)
classifier=classifier.fit(pred_train,tar_train)

```

```python
#plotting a decision tree with depth 5

classifier_2=DecisionTreeClassifier(max_depth=5)
classifier_2=classifier_2.fit(pred_train,tar_train)

dot_data = StringIO()
export_graphviz(classifier_2,out_file=dot_data,filled=True, rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue()) 
Image(graph.create_png())
```

```python
# caluculating the accuracy, sensitivity and specificity 
predictions=classifier.predict(pred_test)
confusion = sklearn.metrics.confusion_matrix(tar_test,predictions)

accuracy = sklearn.metrics.accuracy_score(tar_test, predictions)
sensitivity = confusion[1][1]/float(confusion[1:].sum())
specificity = confusion[0][0]/float(confusion[:,0].sum())
print ("The accuracy is {0:.3f}".format(accuracy))
print ("The sensitivity is {0:.3f}".format(sensitivity))
print ("The specificity is {0:.3f}".format(specificity))
```

```python
# plotting roc curve with AUC
from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(tar_test, predictions)
roc_auc = auc(false_positive_rate, true_positive_rate)

# Plot of a ROC curve for a classification tree
%matplotlib notebook
plt.figure()
plt.plot(false_positive_rate, true_positive_rate, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.05])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```
